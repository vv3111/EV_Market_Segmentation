```python
# importing libraries
import numpy as np
import pandas as pd
```

```python
# loading dataset 
df = pd.read_csv("EV_sales_India (data.gov.in).csv")
df.head()
```

```python
df = df.fillna(0)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
df.plot(kind='bar', x='State Name', y=['Two Wheeler', 'Three Wheeler', 'Four Wheeler', 'Goods Vehicles', 'Public Service Vehicle', 'Special Category Vehicles', 'Ambulance/Hearses', 'Construction Equipment Vehicle', 'Other'], figsize=(10,5))
plt.xlabel("State")
plt.ylabel("Total EV Sales")
```

```python
df.corr()
```

```python
import seaborn as sns
corrMatrix = df.corr() 
plt.gcf().set_size_inches(10,9)
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
for i in df['State Name']:
    state_data = df.loc[df['State Name']==i].iloc[:,1:]
    if 'Grand Total' in state_data.columns:
        state_data = state_data.drop('Grand Total', axis=1)
    maxi = state_data.max().max()
    for j in state_data.columns:
        if state_data.loc[state_data[j]==maxi, j].any():
            print(f"{i}: {j}")
```

```python
# states with most vehicles of any type
df=df.iloc[:-1,:]
for col in df.columns[1:]:
    if col != 'Grand Total':
        print(f"{col}: {df.loc[df[col].idxmax(), 'State Name']}")
```
